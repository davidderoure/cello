## Configure paths

Edit the paths below to match your Google Drive layout.

```python
from pathlib import Path

DRIVE = Path('/content/drive/MyDrive')

# ── Input files (trimmed, silence/talking removed) ──────────────────────────
TRIMMED = DRIVE / 'cello_session' / 'trimmed'

INPUT_FILES = {
    'mono_1': TRIMMED / 'mono_1.wav',
    'mono_2': TRIMMED / 'mono_2.wav',
    'stereo': TRIMMED / 'stereo.wav',
}

# ── Output root (sub-directories created per mic) ───────────────────────────
OUTPUT_ROOT = DRIVE / 'cello_session' / 'samples'

# Verify inputs exist
for name, path in INPUT_FILES.items():
    status = '✓' if path.exists() else '✗ NOT FOUND'
    print(f'  {name}: {path}  {status}')
```

## Optional: override detection thresholds

Leave unchanged to use the defaults.  See the README for guidance on tuning.

```python
from cello_sampler import config

# Onset detection
config.ONSET_THRESHOLD_MULTIPLIER = 1.5   # raise if too many false onsets
config.MIN_NOTE_DURATION_MS       = 50.0

# Attack capture — how much audio before the detected onset is kept.
# Cello bow engagement takes 20–50 ms before the onset detector fires, so
# this must be large enough to include the transient.  Default 20 ms.
# Raise to 50 ms if samples still sound clipped at the start.
config.PRE_ONSET_SAMPLES = int(20 / 1000 * 48_000)   # 20 ms = 960 samples

# Quality gating
config.PITCH_CONFIDENCE_THRESHOLD     = 0.85  # lower if too many low-confidence rejections
config.MAX_INTONATION_DEVIATION_CENTS = 30.0
config.POLYPHONY_SALIENCE_THRESHOLD   = 0.30  # raise if clean notes are wrongly rejected

# Articulation
config.PIZZICATO_MAX_ATTACK_MS   = 15.0
config.STACCATO_MAX_DURATION_MS  = 250.0
config.VIBRATO_MIN_DEPTH_CENTS   = 8.0    # std dev ~8 ¢ = ~23 ¢ peak-to-peak

# Minimum duration for sustaining articulations.
# Shorter takes are rejected (too_short).
config.MIN_LEGATO_DURATION_MS  = 500.0   # ~0.5 s minimum for legato
config.MIN_VIBRATO_DURATION_MS = 500.0   # ~0.5 s minimum

# Workers: keep at 1 in Colab — each extra process re-loads TensorFlow
# and fights for the same GPU, which is slower not faster.
N_WORKERS = 1

print('Thresholds configured.')
```

## Run the pipeline

```python
from cello_sampler.pipeline import run

results = {}

for mic_name, input_path in INPUT_FILES.items():
    if not input_path.exists():
        print(f'Skipping {mic_name}: file not found at {input_path}')
        continue

    output_dir = OUTPUT_ROOT / mic_name
    output_dir.mkdir(parents=True, exist_ok=True)

    print(f'\n── {mic_name} ──────────────────────────────────')
    print(f'   Input  : {input_path}')
    print(f'   Output : {output_dir}')

    result = run(
        input_path=input_path,
        output_dir=output_dir,
        n_workers=N_WORKERS,
    )

    results[mic_name] = result
    print(f'   Accepted : {result.n_accepted}')
    print(f'   Rejected : {result.n_rejected}')

print('\nDone.')
```

## Summary

```python
import csv
from collections import Counter

for mic_name, result in results.items():
    index_path = OUTPUT_ROOT / mic_name / '_index.csv'
    if not index_path.exists():
        continue

    with index_path.open() as f:
        rows = list(csv.DictReader(f))

    articulations = Counter(r['articulation'] for r in rows)
    notes         = Counter(r['note'] for r in rows)

    print(f'\n{mic_name}')
    print(f'  Total accepted : {result.n_accepted}')
    print(f'  Total rejected : {result.n_rejected}')
    print(f'  By articulation: {dict(articulations)}')
    print(f'  Unique pitches : {len(notes)}  ({" ".join(sorted(notes))})')
```

## Add loop points (optional)

Embeds sustain loop markers into legato and vibrato WAV files so a sampler
can loop them indefinitely.

**For legato samples this step is usually not worth it.** Legato cello tone
evolves continuously (bow pressure, natural vibrato, resonance) so any splice
point tends to be audible. The default SFZ mode `no_loop` plays each sample
once in full — no splice, completely natural — which sounds better for legato.
Skip this step and use `no_loop` (the default) unless your recordings are
unusually stable and you specifically need infinite sustain.

If you do run this step, use `--loop-mode loop_sustain` in the SFZ cell below
so the sampler actually uses the embedded markers.

```python
import sys
sys.path.insert(0, '/content/cello')

from add_loop_points import main as add_loops

for mic_name in INPUT_FILES:
    output_dir = OUTPUT_ROOT / mic_name
    if not output_dir.exists():
        print(f'Skipping {mic_name}: output directory not found')
        continue

    print(f'\n── {mic_name} ──────────────────────────────────')

    # Process legato and vibrato folders.
    # Raise --max-splice-error if few files receive loop points.
    add_loops([
        str(output_dir),
        '--min-cycles',       '4',
        '--max-splice-error', '0.05',
    ])

print('\nLoop points done.')
```

## Generate SFZ instrument files

Creates one `.sfz` file per articulation with explicit MIDI key zones, round-robin takes,
and per-note volume normalisation — all takes of the same note are matched to the loudest
take so round-robin playback is level-consistent. WAV files are not modified.

**Loop mode** — the default is `no_loop`: each sample plays once in full, then stops.
This avoids splice artefacts and is recommended for legato cello.  If you ran the
"Add loop points" step above, change `--loop-mode` to `loop_sustain` so the sampler
uses the embedded markers.

Load the `.sfz` files in **sfizz** (free AU plugin: https://sfizz.com) inside Logic Pro
instead of dragging WAV files directly into Logic's Sampler.

```python
import importlib, sys, csv
from pathlib import Path

# Always reload generate_sfz from disk so git-pulled changes take effect
# without needing a full kernel restart.
if 'generate_sfz' in sys.modules:
    importlib.reload(sys.modules['generate_sfz'])

from generate_sfz import main as generate_sfz

# ── Loop mode ────────────────────────────────────────────────────────────────
# "no_loop"      — play each sample once in full (default, recommended for legato)
# "loop_sustain" — loop while key held, play tail on release
#                  (only useful if you ran the "Add loop points" step above)
LOOP_MODE = "no_loop"

for mic_name in INPUT_FILES:
    output_dir = OUTPUT_ROOT / mic_name
    if not output_dir.exists():
        print(f'Skipping {mic_name}: output directory not found')
        continue

    # Quick sanity check: show what articulations are in the index.
    index_path = output_dir / '_index.csv'
    if not index_path.exists():
        print(f'{mic_name}: _index.csv not found in {output_dir}')
        continue
    with index_path.open(newline='', encoding='utf-8') as f:
        rows = list(csv.DictReader(f))
    arts = sorted({r.get('articulation', '').strip() for r in rows} - {''})
    cols = list(rows[0].keys()) if rows else []
    print(f'\n-- {mic_name}  ({len(rows)} rows, articulations: {arts})')
    print(f'   CSV columns: {cols}')

    generate_sfz([str(output_dir), '--loop-mode', LOOP_MODE])

print('\nDownload the .sfz files from Google Drive to your Mac.')
```
